Q1：利用所學的判斷式技能，設計出讓使用者輸入一個正整數後，程式會回傳該正整數為奇數或偶數。**(變數命名可自行定義)**

提示1：該正整數能夠被2整除(也就是餘數為0)就是偶數。

提示2：請先判斷使用者輸入是否為正整數，若是，才開始判斷為奇偶數；若否，則輸出錯誤訊息。(運用巢狀判斷式)

擷取.PNG

```python
#提示使用者輸入一個正整數(運用input技能)
n = int(input("請輸入正整數: "))
#判斷為奇數或偶數(運用巢狀判斷式技能)
if (n >= 0):
  if (n % 2 == 0):
    print("{} 為偶數! ".format(n))
  else:
    print("{} 為偶數! ".format(n))
else:
  print("您輸入非正整數! ")





```

Q2：我們知道1～3月為春天；4～6月為夏天；7～9月為秋天；10～12月為冬天。請設計程式，讓使用者輸入月份後，請輸出該月份為哪個季節。若輸入非上述月份，請輸出錯誤結果。**(變數命名可自行定義)**

擷取2.PNG

```python
#提示使用者輸入月份(運用input技能)
month = int(input("請輸入月份(阿拉伯數字): "))
#判斷季節(運用巢狀判斷式技能)
if (month>=1 and month<=12):
  if (month==1 or month==2 or month==3):
    print("%d 月是春天!" % month)
  elif (month==4 or month==5 or month==6):
    print("%d 月是夏天!" % month)
  elif (month==7 or month==8 or month==9):
    print("%d 月是秋天!" % month)
  else:
    print("%d 月是冬天!" % month)
else:
  print("輸入錯誤! %d 月不在範圍內!" % month)










```
